```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints=np.array([10,5])
ypoints=np.array([30,60])
plt.plot(xpoints,ypoints)
plt.show
```

```python
plt.plot([3,50],[5,30])
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.scatter(x,y)
plt.show()
```

```python
#plotting without y-points
ypoints=np.array([3,4,5,6,7])
plt.plot(ypoints)
plt.show()
```

```python
ypoints=np.array([3,4,5,6,7])
plt.plot(ypoints,marker='*',mec='r')
plt.show()
```

```python
#plotting without x-points
ypoints=np.array([3,4,5,6,7])
plt.plot(ypoints,marker='o',mec='b',color='r',linewidth='20')
plt.show()
```
