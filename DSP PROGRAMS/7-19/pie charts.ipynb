```python
import matplotlib.pyplot as plt

country=['US','india','germany','UK','france']
GDP_per_capita=[480,380,500,450,440]

plt.bar(country,GDP_per_capita,color='gray')
plt.title('country vs GDP per capita',color='green')
plt.xlabel('country')
plt.ylabel('GDP per capita')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,10,10,0,0])
y=np.array([0,0,10,10,0])
x1=np.array([0,5,10])
y1=np.array([10,25,10])
x3=np.array([2,2,4,4,6,6,8,8])
y3=np.array([0,10,10,0,0,10,10,0])
x4=np.array([1.7,8.3,6.7,3.3])
y4=np.array([15,15,20,20])
plt.plot(x,y,marker="o",color="r")
plt.plot(x1,y1,marker="o",color="b",mec="b")
plt.plot(x3,y3,marker="o",color="b",mec="b")
plt.plot(x4,y4,marker="o",color="b",mec="b")
plt.grid()
plt.show()
```
