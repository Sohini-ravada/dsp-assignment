```python
import matplotlib.pyplot as plt

country=['US','india','germany','UK','france']
GDP_per_capita=[480,380,500,450,440]

plt.bar(country,GDP_per_capita,color='hotpink')
plt.title('country vs GDP per capita',color='green')
plt.xlabel('country')
plt.ylabel('GDP per capita')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
sub=np.array(["dsp","cd","wt","coa"])
marks=np.array([30,35,47,68])
plt.barh(sub,marks,color="yellow")
plt.xlabel("sub")
plt.ylabel("marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,25])
plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["dsp","cd","wt"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,30,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["dsp","co","wt"])
plt.legend(title="subects")
plt.show()
```
